```python
# 0. import matplotpib and gt it ready for plotting in Jupyter
import matplotlib.pyplot as plt

# 1. 
x = [1, 2, 3, 4]
y = [11, 22, 33, 55]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10))

# 3. Plot the data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="simple Plot", 
       xlabel="xaxis",
       ylabel="yaxis")
# 5. Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
plt.show();
```

## Making Figures with NumPy Arrays
We want:
* Line Plot
* Scatter Plot
* Bar Plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
# create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2);
plt.show();
```

```python
# use same datat to make scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
plt.show();
```

```python
fir, ax = plt.subplots()
ax.scatter(x, np.sin(x));
plt.show();
```

```python
# make a plot from dictionary
nut_butter_prices = {"almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title="Tom's Nut Butter Store",
       ylabel="Price ($)");
plt.show();
```

```python
fig, ax = plt.subplots();
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
plt.show();
```

```python
# Make some data for histograms and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
plt.show();
```

# Two options for subplots

```python
# Subplot option 1 (recommended for rest of matplotlib)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5));
# plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(19000));
plt.show();
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))
# plot to each diff index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
plt.show();
```
